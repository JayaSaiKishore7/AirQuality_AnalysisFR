```python
##Train Test Split Series 
```

```python
# Sort chronologically
df_sample = df_sample.sort_values("date")

# Define target variable
target = "valeur"

# Define feature list
features = [
    "Latitude", "Longitude",
    "hour", "day", "month", "year", "weekday", "is_weekend",
    "Polluant_encoded", "influence_encoded", "evaluation_encoded",
    "implantation_encoded", "site_encoded",
    "lag_1", "lag_24", "rolling_3"
]

# Create X and y
X = df_sample[features]
y = df_sample[target]

# Split index for 80% train / 20% test
split_index = int(len(df_sample) * 0.8)

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

X_train.shape, X_test.shape
```

```python
##Train Random Forest Model
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
## Train XGBoost Model
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
```

```python
##Evaluate Both Models 
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate(true, pred):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    r2 = r2_score(true, pred)
    return mae, rmse, r2

rf_mae, rf_rmse, rf_r2 = evaluate(y_test, rf_pred)
xgb_mae, xgb_rmse, xgb_r2 = evaluate(y_test, xgb_pred)

print("RandomForest  → MAE:", rf_mae, " RMSE:", rf_rmse, " R²:", rf_r2)
print("XGBoost       → MAE:", xgb_mae, " RMSE:", xgb_rmse, " R²:", xgb_r2)
```

```python
## Picking the Best Model
```

```python
best_model = rf_model if rf_rmse < xgb_rmse else xgb_model
best_name = "RandomForest" if rf_rmse < xgb_rmse else "XGBoost"

print("\nBest Model Selected:", best_name)
```

```python
##Next 24-Hour Forecast
```

```python
import pandas as pd

def forecast_next_24(df, model):
    """
    Forecast next 24 hours using last row of df as starting point.
    df must be sorted by date and contain the final feature columns.
    """
    df = df.sort_values("date")
    
    # Start from last known row
    last_row = df.iloc[-1:].copy()
    
    feature_cols = [
        "Latitude", "Longitude",
        "hour", "day", "month", "year", "weekday", "is_weekend",
        "Polluant_encoded", "influence_encoded", "evaluation_encoded",
        "implantation_encoded", "site_encoded",
        "lag_1", "lag_24", "rolling_3"
    ]
    
    forecasts = []
    timestamps = []

    for step in range(1, 25):  # 24 hours ahead
        X_last = last_row[feature_cols]
        
        # 1) predict next value
        next_val = model.predict(X_last)[0]
        forecasts.append(next_val)
        
        # 2) create next timestamp
        next_time = last_row["date"].iloc[0] + pd.Timedelta(hours=1)
        timestamps.append(next_time)
        
        # 3) build new row for next iteration
        new_row = last_row.copy()
        new_row["date"] = next_time
        
        # update time features
        new_row["hour"] = next_time.hour
        new_row["day"] = next_time.day
        new_row["month"] = next_time.month
        new_row["year"] = next_time.year
        new_row["weekday"] = next_time.weekday()
        new_row["is_weekend"] = 1 if new_row["weekday"].iloc[0] >= 5 else 0
        
        # update lag and rolling features
        prev_val = last_row["valeur"].iloc[0]  # last known/forecast value
        new_row["lag_1"] = prev_val
        # lag_24: for simplicity we keep the old lag_24 (you could build a proper buffer if needed)
        # update rolling_3 as exponential moving-style
        new_row["rolling_3"] = (last_row["rolling_3"] * 2 + next_val) / 3
        
        # update the 'valeur' column to the new predicted value
        new_row["valeur"] = next_val
        
        # feed it into next iteration
        last_row = new_row.copy()
    
    forecast_df = pd.DataFrame({
        "forecast_time": timestamps,
        "forecast_value": forecasts
    })
    
    return forecast_df
```

```python
df_sample = df_sample.sort_values("date")

forecast_df = forecast_next_24(df_sample, best_model)
forecast_df.head()
```
