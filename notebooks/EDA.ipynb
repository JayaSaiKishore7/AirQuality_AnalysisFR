```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

models = {}
scores = {}

for p in pollutants:
    print(f"\nTraining model for pollutant: {p}")
    
    df_p = df_sample[df_sample["Polluant_encoded"] == p].copy()
    
    # Split chronological
    split_i = int(len(df_p) * 0.8)
    
    X_p = df_p.drop(columns=["valeur"])
    y_p = df_p["valeur"]
    
    X_train_p = X_p.iloc[:split_i]
    X_test_p  = X_p.iloc[split_i:]
    y_train_p = y_p.iloc[:split_i]
    y_test_p  = y_p.iloc[split_i:]
    
    # Train model
    model_p = XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=8,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    
    model_p.fit(X_train_p, y_train_p)
    y_pred_p = model_p.predict(X_test_p)
    
    # Evaluate
    mae = mean_absolute_error(y_test_p, y_pred_p)
    rmse = np.sqrt(mean_squared_error(y_test_p, y_pred_p))
    r2   = r2_score(y_test_p, y_pred_p)
    
    models[p] = model_p
    scores[p] = {"MAE": mae, "RMSE": rmse, "R2": r2}
    
    print(f"Pollutant {p}: MAE={mae:.3f}, RMSE={rmse:.3f}, R²={r2:.4f}")
```

```python
import pandas as pd

next_hour_forecasts = {}

for p in pollutants:
    # Filter data for this pollutant and sort in time order
    df_p = df_sample[df_sample["Polluant_encoded"] == p].copy()
    df_p = df_p.sort_values(["year", "month", "day", "hour"])
    
    # Take the last known row (most recent hour)
    last_row = df_p.iloc[-1:].copy()
    
    # Drop target column to get features only
    X_last = last_row.drop(columns=["valeur"], errors="ignore")
    
    # Predict next hour pollution for this pollutant
    y_next = models[p].predict(X_last)[0]
    
    next_hour_forecasts[p] = y_next

next_hour_forecasts
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
from lightgbm import LGBMRegressor
use_lightgbm = True
use_lightgbm = False
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    return mae, rmse, r2
```

```python
models_to_test = {
    "XGBoost": XGBRegressor(
        n_estimators=300, learning_rate=0.05,
        max_depth=8, subsample=0.8, colsample_bytree=0.8,
        random_state=42
    ),

    "RandomForest": RandomForestRegressor(
        n_estimators=300, max_depth=20,
        random_state=42, n_jobs=-1
    )
}

if use_lightgbm:
    models_to_test["LightGBM"] = LGBMRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=-1,
        random_state=42
    )
```

```python
results = {}

for name, model in models_to_test.items():
    print(f"\nTraining {name}...")
    mae, rmse, r2 = evaluate_model(model, X_train, y_train, X_test, y_test)
    results[name] = {"MAE": mae, "RMSE": rmse, "R2": r2}
    print(f"{name}: MAE={mae:.4f}, RMSE={rmse:.4f}, R²={r2:.4f}")
```
