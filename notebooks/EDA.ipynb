```python
##HYPOTHESIS TESTING
```

```python
from scipy import stats
import numpy as np

# Make valeur numeric
#df['valeur'] = pd.to_numeric(df['valeur'], errors='coerce')
#df.loc[df['valeur'] < 0, 'valeur'] = np.nan  # remove invalid values

# Two groups
urban = df_sample[df_sample["type d'implantation"]=="Urbaine"]['valeur'].dropna()
peri = df_sample[df_sample["type d'implantation"]=="Périurbaine"]['valeur'].dropna()

# Hypothesis test
t_stat, p_val = stats.ttest_ind(urban, peri, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_val)
```

```python
pm10 = df_sample[df_sample["Polluant"]=="PM10"]["valeur"].dropna()
o3 = df_sample[df_sample["Polluant"]=="O3"]["valeur"].dropna()

t_stat, p_val = stats.ttest_ind(pm10, o3, equal_var=False)
print(p_val)
```

```python
##FEATURE ENGINEERING 
```

```python
df_sample["date"] = pd.to_datetime(df_sample["Date de début"])
```

```python
#TIME FEATURES EXTRACION
df_sample["hour"] = df_sample["date"].dt.hour
df_sample["day"] = df_sample["date"].dt.day
df_sample["month"] = df_sample["date"].dt.month
df_sample["year"] = df_sample["date"].dt.year
df_sample["weekday"] = df_sample["date"].dt.weekday     # 0=Mon, 6=Sun
df_sample["is_weekend"] = df_sample["weekday"].isin([5,6]).astype(int)

df_sample["week_of_year"] = df_sample["date"].dt.isocalendar().week
df_sample["day_of_year"] = df_sample["date"].dt.dayofyear
```

```python
#SEASON EXTRACTION
def get_season(month):
    if month in [12,1,2]:
        return "winter"
    elif month in [3,4,5]:
        return "spring"
    elif month in [6,7,8]:
        return "summer"
    else:
        return "autumn"

df_sample["season"] = df_sample["month"].apply(get_season)
df_sample["season"]
```
