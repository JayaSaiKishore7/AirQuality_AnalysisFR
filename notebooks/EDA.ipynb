```python
# 1. Drop useless columns
columns_to_drop = [
    "valeur brute", 
    "Réglementaire", 
    "type de valeur",
    "taux de saisie",
    "couverture temporelle",
    "couverture de données",
    "validité"
]
df_clean = df_raw.drop(columns=columns_to_drop)

# 2. Remove missing pollution values
df_clean = df_clean.dropna(subset=["valeur"])

# 3. Remove zero values
df_clean = df_clean[df_clean["valeur"] != 0]

# 4. Convert dates
df_clean["Date de début"] = pd.to_datetime(df_clean["Date de début"], errors="coerce")
df_clean["Date de fin"]   = pd.to_datetime(df_clean["Date de fin"], errors="coerce")

# 5. Remove invalid dates
df_clean = df_clean.dropna(subset=["Date de début", "Date de fin"])
df_clean = df_clean[df_clean["Date de fin"] >= df_clean["Date de début"]]

# 6. Remove duplicates
df_clean = df_clean.drop_duplicates()

# 7. Remove negative values
df_clean = df_clean[df_clean["valeur"] >= 0]
```

```python
df_clean.to_csv("../data/processed/AirFrance_2025_cleaned.csv", index=False)
```
