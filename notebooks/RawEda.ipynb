```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Select only the numerical columns of interest for the pairplot
numerical_cols = ['valeur', 'valeur brute', 'Latitude', 'Longitude']
sns.pairplot(df_raw[numerical_cols])
plt.show()
```

```python
## The pairplot shows that valeur and valeur brute are almost perfectly correlated.so, we can remove one of them for modeling.
## Latitude and Longitude form clear clusters and we can see there are fixed plots in the data.
## Pollution values are strongly right-skewed, with many low readings and a few high spikes.
##There is no clear linear relationship between pollution and location.
```

```python
sns.histplot(df_raw["valeur"], bins=50, kde=True)
plt.show()
```

```python
## Here we can see that the pollution values are right-skewed with a long tail on the right side. we can use log transformation to reduce the skewness
```

```python
df_raw["valeur"].skew()
```

```python
## Here we reduced the skewness significantly using log transformation
```

```python
df_raw["valeur_log"] = np.log1p(df_raw["valeur"])
df_raw["valeur_log"].skew()
```

```python
## Looking at the boxplot of the original valeur column and the log-tranformed valeur_log column to see the effect on outliers 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.boxplot(x=df_raw["valeur"])
plt.title("Boxplot of valeur (Pollutant Values)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.boxplot(x=df_raw["valeur_log"])
plt.title("Boxplot of valeur (Pollutant Values)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = df_raw.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
